```python
display(df)
b=df.dropna()
b
```

```python
b.info()
```

```python
b['Quantity Ordered'] = pd.to_numeric(b['Quantity Ordered'])
b['Price Each'] = pd.to_numeric(b['Price Each'])
b
```

```python
b.info()
```

```python
b['month']= pd.DatetimeIndex(b['Order Date']).month
b['month'] = b['month'].astype(int)
b
```

```python
b['city'] = b['Purchase Address'].apply(lambda x: x.split(",")[1])
b
```

```python
from scipy.stats import skew, kurtosis
datA= np.array(b['Quantity Ordered'])
datAsum = np.sum(datA)
datAmean = np.mean(datA)

datAstd = np.std(datA)
datAskew= skew(datA)
datAkur = kurtosis(datA)
print(datAsum, datAmean, datAstd,datAskew,datAkur)
```

```python
import seaborn as sns
coorelation=b.corr()
plt.imshow(coorelation)
print(coorelation)
sns.heatmap(coorelation)
```

```python
b['sale'] = b['Quantity Ordered'] *b['Price Each']
b
```

```python
saleinmonth=b.groupby('month')['sale'].sum()
saleinmonth
plt.bar(saleinmonth.index,saleinmonth.values)
plt.xlabel("sales")
plt.ylabel("month")
plt.xticks(range(1,13))
plt.show()
```

```python
citysold = b.groupby("city")['Quantity Ordered'].sum()
citysold
```

```python
citysold = b.groupby("city")['Quantity Ordered'].sum()
plt.bar(citysold.index, citysold.values, label="city")
plt.xlabel("city")
plt.ylabel("Qunatity Ordered")
plt.legend()
plt.xticks(rotation=90)


```
